```python
import numpy as np
import pandas as pd
T = 0.9285
A = 1119
WD = 0.30
WL = 0.25
WPP = 1.00 
Peso_piso = A * (WD + WL + WPP)
k = 0.75 + 0.50 * T
V = 4190.53

h_pisos = np.array([4, 7.6, 11.2, 14.8, 18.4, 22.0, 25.6, 29.2, 32.8, 36.4, 40.0, 43.6, 47.2])
h_pisos2 = np.array([4, 3.6, 3.6, 3.6, 3.6, 3.6, 3.6, 3.6, 3.6, 3.6, 3.6, 3.6, 3.6])
Pesos = np.ones([13])*Peso_piso

sum = np.sum(Pesos * (h_pisos ** k))
fx = []

for i in np.arange(0, 13, 1):
    fx.append(Pesos[i]*h_pisos[i]**k / sum * V)

fx = np.array(fx)
fx_col = fx * 0.25
fx_wall = fx * 0.75
fx_df = pd.DataFrame(fx, columns=['fuerzas laterales de piso (T)']) 
fx_col_df = pd.DataFrame(fx_col, columns=['fuerzas laterales de piso (T), para Columnas']) 
fx_wall_df = pd.DataFrame(fx_wall, columns=['fuerzas laterales de piso (T), para Muros']) 

VD = np.sum(fx)
print("La fuerza lateral total VD es: ", VD, "T")

Resul = pd.concat([fx_df, fx_col_df, fx_wall_df], axis=1)
Resul.head(len(h_pisos))
```
